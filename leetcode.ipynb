# 句子相似性

```python
sentence1 = "TjZ ScAi m xz PNWaKigqqY p IyJ B rok"
sentence2 = "TjZ ScAi m LlbJhCf gL u m R pZpiH mSk a og xz PNWaKigqqY p IyJ B rok"
```

```python
base,comp = sentence1.split(),sentence2.split()
if len(base) < len(comp):
    base,comp = comp,base

# 先判断做匹配

i = 0
while i < len(comp) and base[i] == comp[i]:
    i+=1
if i == len(comp):
    print('True')
    
j = 0
while j >= 0 and base[len(base) - j - 1] == comp[len(comp) - j - 1]:
    j += 1
# 同时判断两端匹配
if j == len(comp) or i + j == len(comp):
    print('True')
# 其他场景都是错误的
print('False') 
```

# 完成一半题目问题

```python
questions = [1,5,1,3,4,5,2,5,3,3,8,6]
```

```python
leetcoder = len(questions)/2
# 列表去重
formatList = list(set(questions))
temp_lsit =[]
for x in formatList:
    t=questions.count(x)
    temp_lsit.append(t)
temp_lsit.sort()
temp_lsit.reverse()
print(temp_lsit)
```

```python
leiji = 0
count = 0
for i in range(len(temp_lsit)):
    leiji = leiji + temp_lsit[i]
    count +=1
    if leiji >= leetcoder:
        break
    else:
        continue
```

# 最大的团队表现值

```python
n = 3
speed = [2,8,2]
efficiency = [2,7,1]
ki = 2
```

```python
new_list =[]
for i in range(n):
    new_list.append(i)
print(new_list)
```

```python
new_2_list = []
for k in range(2 ** n):#子集的个数
    combo = []
    for j in range(n):#用来判断二进制数的下标为j的位置的数是否为1
        if (k >> j) % 2:
            combo.append(new_list[j])
    if len(combo) <= ki:
        new_2_list.append(combo)
print(new_2_list)
```

```python
performance_lsit = []
for t in new_2_list:
    print(t)
    speed_2 = 0
    effi_list = []
    for t2 in t:
        index = int(t2)
        speed_2 = speed_2+speed[index]
        effi_list.append(efficiency[index])
    e = min(effi_list)
    p = speed_2 * e
    print(speed_2)
    print(effi_list)
    performance_lsit.append(p)
print(performance_lsit)
print(max(performance_lsit))
```

# 使结果不超过阈值的最小除数

```python
nums = [1,2,5,9]
threshold = 6
```

```python
for i in range(1,max(nums)):
    print(i)
    n_list = []
    for n in nums:
        t = n/i
        if int(t)<t:
            t = int(t+1)
        else:
            t =int(t)
        n_list.append(t)
    t2=0
    for m in n_list:
        t2 =t2+m
    if t2<=threshold:
        print('result:',i)
        break
```
